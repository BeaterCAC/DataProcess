# 5. 新变量生成

## 5.1 家庭占比变量

### 由于三个表格拼接的时会因为id的匹配掉部分数据，为保证计算真实性，用的原始ind数据进行计算的。

```python
ind.replace(['.d','.r','.e','.n'],np.nan,inplace=True)
nvar = ['a2005','a2003', 'a2025b', 'a3106', 'f1001a', 'f2001a', 'f3001', 'f4001']
for i in nvar:
        ind[i] = ind[i].apply(lambda x:float(x) if not pd.isnull(x) else x)
```

```python
tmp = ind.groupby('hhid')['pline'].count().reset_index().rename({'pline':'家庭规模'},axis=1)
laonian = ind.loc[(ind['a2005']<=1957),:].groupby(['hhid'])['a2005'].count().reset_index().rename({'a2005':'家中老人数量'},axis=1)
xiaohai = ind.loc[(ind['a2005']>=1999),:].groupby(['hhid'])['a2005'].count().reset_index().rename({'a2005':'家中小孩数量'},axis=1)
nanxing = ind.loc[(ind['a2003'] == 1),:].groupby(['hhid'])['a2003'].count().reset_index().rename({'a2003':'男性数量'},axis=1)
jiankang = ind.loc[(ind['a2025b']>= 4 ),:].groupby(['hhid'])['a2025b'].count().reset_index().rename({'a2025b':'不健康人数'},axis=1)
zhengfu = ind.loc[(ind['a3106'] == 1 ),:].groupby(['hhid'])['a3106'].count().reset_index().rename({'a3106':'政府工作人员'},axis=1)
yanglao = ind.loc[(ind['f1001a'] != 7788 )&(pd.notna(ind['f1001a'])),:].groupby(['hhid'])['f1001a'].count().reset_index().rename({'f1001a':'养老保险量'},axis=1)
yiliao = ind.loc[(ind['f2001a'] != 7788 )&(pd.notna(ind['f2001a'])),:].groupby(['hhid'])['f2001a'].count().reset_index().rename({'f2001a':'医疗保险量'},axis=1)
shiye = ind.loc[(ind['f3001'] == 1 )&(pd.notna(ind['f3001'])),:].groupby(['hhid'])['f3001'].count().reset_index().rename({'f3001':'失业保险量'},axis=1)
gongjijin = ind.loc[(ind['f4001'] == 1 )&(pd.notna(ind['f4001'])),:].groupby(['hhid'])['f4001'].count().reset_index().rename({'f4001':'住房公积金量'},axis=1)
```

```python
listdata = [laonian,xiaohai, nanxing, jiankang, zhengfu, yanglao, yiliao, shiye, gongjijin]
for i in listdata:
    tmp = pd.merge(tmp,i,left_on='hhid',right_on='hhid',how='left')
tmp.fillna(0,inplace=True)

zhanbilist = [ '家中老人数量','家中小孩数量','男性数量','不健康人数','政府工作人员','养老保险量','医疗保险量','失业保险量','住房公积金量']
for i in zhanbilist:
    tmp[i+'占比'] = tmp[i]/tmp['家庭规模']
```

```python
df12 = pd.merge(df11,tmp,left_on='hhid',right_on='hhid',how='left')
```

## 5.3 其他变量生成

```python
#其他变量
df13['年龄'] = 2019-df13['年龄']
df13['受教育年限'] = df13['文化程度'].replace({1:0,2:6,3:9,4:12,5:12,6:15,7:16,8:18,9:22})
df13['转移收入'] = df13['父母转移收入'] + df13['父母2转移收入'] + df13['其他转移收入'] 
```

```python
# 资产划分和占比
fengxian = df13[['股票', '基金', '金融理财',  '互联网理财', '债券', '衍生品', '非人民币资产', '黄金', '其他金融资产']]
cunkuan = df13[['活期存款', '定期存款']]
df13['风险'] = [1 if x >=1 else 0 for x in list(sum(fengxian.T.values)) ]
df13['存款'] = [1 if x >=1 else 0 for x in list(sum(cunkuan.T.values)) ]

df13['只风险'] = [1 if (df13['风险'][i]==1)&(df13['存款'][i] !=1) else 0 for i in range(len(df13['风险']))]
df13['只存款'] = [1 if (df13['风险'][i]!=1)&(df13['存款'][i] ==1) else 0 for i in range(len(df13['风险']))]
df13['均持有'] = [1 if (df13['风险'][i]==1)&(df13['存款'][i] ==1) else 0 for i in range(len(df13['风险']))]
df13['均不持有'] = [1 if (df13['风险'][i]!=1)&(df13['存款'][i] !=1) else 0 for i in range(len(df13['风险']))]


df13['金融资产总量tmp'] = 0
for i in ['活期存款量',  '定期存款量', 
    '股票量', '基金量',  '互联网理财量',  '金融理财量', '债券量', '衍生品量',  '非人民币资产量', '黄金量',  '其他金融资产量']:
    df13['金融资产总量tmp'] = df13['金融资产总量tmp'] + df13[i]

df13['风险金融资产总量tmp'] = 0
for i in [ '股票量', '基金量',  '互联网理财量',  '金融理财量', '债券量', '衍生品量',  '非人民币资产量', '黄金量',  '其他金融资产量']:
    df13['风险金融资产总量tmp'] = df13['风险金融资产总量tmp'] + df13[i]
df13['风险金融资产量占比'] = df13['风险金融资产总量tmp']/ df13['金融资产总量tmp']

df13['存款金融资产总量tmp'] = 0
for i in [ '活期存款量',  '定期存款量']:
    df13['存款金融资产总量tmp'] = df13['存款金融资产总量tmp'] + df13[i]
df13['存款金融资产量占比'] = df13['存款金融资产总量tmp']/ df13['金融资产总量tmp']

df13.fillna(0,inplace=True)
```
